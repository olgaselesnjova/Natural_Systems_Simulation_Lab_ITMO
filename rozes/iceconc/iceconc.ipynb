```python
import pandas as pd
import numpy as np
import datetime
from matplotlib import dates
from matplotlib import pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
#from matplotlib.ticker import FormatStrFormatter
```

```python
path1 = 'D:/ИТМО/Python/rozes/iceconc/mod01__18-10-18_iceconc.csv' # Добавить файл для новой ПАБС
data1 = pd.read_csv(path1, sep=',',decimal='.') 
df1 = pd.DataFrame(data1)
main1 = df1.iloc[:,1:]

path2 = 'D:/ИТМО/Python/rozes/iceconc/mod01__19-10-13_iceconc.csv' # Добавить файл для новой ПАБС
data2 = pd.read_csv(path2, sep=',',decimal='.') 
df2 = pd.DataFrame(data2)
main2 = df2.iloc[:,1:]

path3 = 'D:/ИТМО/Python/rozes/iceconc/mod01__20-10-16_iceconc.csv' # Добавить файл для новой ПАБС
data3 = pd.read_csv(path3, sep=',',decimal='.') 
df3 = pd.DataFrame(data3)
main3 = df3.iloc[:,1:]
```

```python
picture_name = "PABS_1"
df1_name = "PABS_1_1"
df2_name = "PABS_1_2"
df3_name = "PABS_1_3"

x1 = main1.iloc[:,0]
x1 = pd.to_datetime(x1, format='%Y-%m-%d %H:%M:%S', errors='ignore')

x2 = main2.iloc[:,0]
x2 = pd.to_datetime(x2, format='%Y-%m-%d %H:%M:%S', errors='ignore')

x3 = main3.iloc[:,0]
x3 = pd.to_datetime(x3, format='%Y-%m-%d %H:%M:%S', errors='ignore')

y1 = main1.iloc[:,1]
y2 = main2.iloc[:,1]
y3 = main3.iloc[:,1]

fig, ax = plt.subplots(figsize=(20, 10))

#ax.yaxis.set_major_formatter(FormatStrFormatter('%.4f'))

ax.plot(x1, y1, color="g", label= df1_name, marker = '.')  
plt.plot(x2, y2, color='r', label= df2_name, marker = '.')   
plt.plot(x3, y3, color='b', label= df3_name, marker = '.')   

plt.gcf().autofmt_xdate(rotation=90)
#ax.xaxis.set_major_locator(dates.DayLocator(interval=10))  # for days
ax.xaxis.set_major_locator(dates.MonthLocator(interval=1, bymonthday=1))   # for month revert bymonthday=-1
plt.xticks(rotation=45, ha='right')  

plt.xlabel("Time", fontsize=12)   
plt.ylabel((f'iceconc (NEMO)'), fontsize=12)    
plt.title(picture_name, fontsize=14)                                 
plt.grid()
plt.legend()
plt.savefig(f'D:/ИТМО/Python/rozes/iceconc/{picture_name}') 
plt.show()   
```
